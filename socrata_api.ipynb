```python
hub_311 = []
for page_num in range(1,14):
    url311 = 'https://data.nashville.gov/resource/7qhx-rexh.json'
    params = {'$limit':50000,
              '$offset':((page_num-1)*50000) 
         }
    request311 = requests.get(url311, params)
    json311 = request311.json()
    hub_311.extend(json311)
```
